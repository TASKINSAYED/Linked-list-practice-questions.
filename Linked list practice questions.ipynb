```python
#3)
class ListNode:
    def __init__(self, value=0):
        self.value = value
        self.next = None

def has_cycle(head):
    if not head or not head.next:
        return False
    
    slow = head
    fast = head.next
    
    while fast and fast.next:
        if slow == fast:
            return True
        slow = slow.next
        fast = fast.next.next
    
    return False

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = head.next 

print(has_cycle(head)) 

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

print(has_cycle(head)) 
```

```python
#4)
class ListNode:
    def __init__(self, value=0):
        self.value = value
        self.next = None

def merge_sorted_lists(head1, head2):
    dummy = ListNode()
    current = dummy

    while head1 and head2:
        if head1.value < head2.value:
            current.next = head1
            head1 = head1.next
        else:
            current.next = head2
            head2 = head2.next
        current = current.next

    if head1:
        current.next = head1
    if head2:
        current.next = head2

    return dummy.next

list1 = ListNode(1)
list1.next = ListNode(3)
list1.next.next = ListNode(5)

list2 = ListNode(2)
list2.next = ListNode(4)
list2.next.next = ListNode(6)

merged_head = merge_sorted_lists(list1, list2)

current = merged_head
while current:
    print(current.value, end=" -> ")
    current = current.next
print("None")
```

```python
#5)
class ListNode:
    def __init__(self, value=0):
        self.value = value
        self.next = None

def remove_nth_from_end(head, n):
    if not head or n <= 0:
        return head
    
    dummy = ListNode()
    dummy.next = head
    
    fast = dummy
    slow = dummy
    
    for _ in range(n):
        fast = fast.next
        if not fast:
            return head 
    
    while fast.next:
        slow = slow.next
        fast = fast.next
    
    
    slow.next = slow.next.next
    
    return dummy.next 

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)
head.next.next.next.next.next = ListNode(6)

head = remove_nth_from_end(head, 2)
current = head
while current:
    print(current.value, end=" -> ")
    current = current.next
print("None")
```

```python
#6)
class ListNode:
    def __init__(self, value=0):
        self.value = value
        self.next = None

def remove_duplicates(head):
    current = head

    while current and current.next:
        if current.value == current.next.value:
            current.next = current.next.next
        else:
            current = current.next

    return head
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(3)
head.next.next.next.next = ListNode(4)
head.next.next.next.next.next = ListNode(4)
head.next.next.next.next.next.next = ListNode(4)
head.next.next.next.next.next.next.next = ListNode(5)


head = remove_duplicates(head)

current = head
while current:
    print(current.value, end=" -> ")
    current = current.next
print("None")
```

```python
#7)
class ListNode:
    def __init__(self, value=0):
        self.value = value
        self.next = None

def get_length(head):
    length = 0
    current = head
    while current:
        length += 1
        current = current.next
    return length

def get_intersection(head1, head2):
    len1 = get_length(head1)
    len2 = get_length(head2)
    while len1 > len2:
        head1 = head1.next
        len1 -= 1
    while len2 > len1:
        head2 = head2.next
        len2 -= 1
    while head1 != head2:
        head1 = head1.next
        head2 = head2.next

    return head1  
list1 = ListNode(1)
list1.next = ListNode(2)
list1.next.next = ListNode(3)
list1.next.next.next = ListNode(4)
intersection_node = ListNode(8)
list1.next.next.next.next = intersection_node
list1.next.next.next.next.next = ListNode(6)
list1.next.next.next.next.next.next = ListNode(9)

list2 = ListNode(5)
list2.next = intersection_node
list2.next.next = ListNode(6)
list2.next.next.next = ListNode(7)

intersection = get_intersection(list1, list2)

if intersection:
    print("Intersection found at node with value:", intersection.value)
else:
    print("No intersection found")
```

```python
#8)
class ListNode:
    def __init__(self, value=0):
        self.value = value
        self.next = None

def rotate_right(head, k):
    if not head or k == 0:
        return head
    length = 1
    last_node = head
    while last_node.next:
        last_node = last_node.next
        length += 1
    k = k % length
    
    if k == 0:
        return head
    new_head_prev = head
    for _ in range(length - k - 1):
        new_head_prev = new_head_prev.next

    new_head = new_head_prev.next
    new_head_prev.next = None
    last_node.next = head
    
    return new_head

head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(8)
head.next.next.next.next.next = ListNode(6)
head.next.next.next.next.next.next = ListNode(9)

head = rotate_right(head, 2)
current = head
while current:
    print(current.value, end=" -> ")
    current = current.next
print("None")
```

```python
#9)
class ListNode:
    def __init__(self, value=0):
        self.value = value
        self.next = None

def add_two_numbers(l1, l2):
    dummy_head = ListNode()
    current = dummy_head
    carry = 0

    while l1 or l2:
        val1 = l1.value if l1 else 0
        val2 = l2.value if l2 else 0
        total = val1 + val2 + carry
        carry = total // 10
        current.next = ListNode(total % 10)
        current = current.next
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next
    if carry:
        current.next = ListNode(carry)

    return dummy_head.next
num1 = ListNode(2)
num1.next = ListNode(4)
num1.next.next = ListNode(3)

num2 = ListNode(5)
num2.next = ListNode(6)
num2.next.next = ListNode(4)

result = add_two_numbers(num1, num2)

while result:
    print(result.value, end=" -> ")
    result = result.next
print("None")
```

```python
#10)
class ListNode:
    def __init__(self, value=0, next=None, random=None):
        self.value = value
        self.next = next
        self.random = random

def clone_linked_list(head):
    if not head:
        return None
    mapping = {}
    current = head
    while current:
        mapping[current] = ListNode(current.value)
        current = current.next
    current = head
    while current:
        cloned_node = mapping[current]
        cloned_node.next = mapping.get(current.next)
        cloned_node.random = mapping.get(current.random)
        current = current.next
    return mapping[head]
node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node5 = ListNode(5)

node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

node1.random = node3
node2.random = node5
node3.random = node2
node4.random = node1
node5.random = node4

cloned_head = clone_linked_list(node1)

current = cloned_head
while current:
    print("Value:", current.value, end=", ")
    if current.next:
        print("Next:", current.next.value, end=", ")
    if current.random:
        print("Random:", current.random.value, end="")
    print()
    current = current.next
```
